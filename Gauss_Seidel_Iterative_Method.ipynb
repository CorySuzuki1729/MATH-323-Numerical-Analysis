```python
# 7.2 Gauss-Seidel Iterative Method
# To solve Ax = b given an initial approximation x0:
# Input: the number of equations and unknowns n;
#        the entries a[i,j], 1 <= i, j <= n of the matrix A;
#        the entries b[i], 1 <= i <= n of b; the entries XO[i],
#        of XO = x0; tolerance TOL; maximum number of iterations N.
# Output: the approximate solution x[1],...,x[n] or a message that the
#         number of iterations was exceeded.

import numpy as np
```

```python
def inp(flag, n, A, b, x_0, tol, N):
    flag = False
    while not flag:
        n = int(input("Please input the number of unknowns and equations (n): "))
        if n <= 0:
            print("Must be a positive integer value.")
        else:
            flag = True
    
    A = np.empty([n, n])
    b = np.empty(n)
    x_0 = np.empty(n)

    flag = False
    while not flag:
        print("Please enter the coefficients for matrix A: ")
        for i in range(n):
            print(f"Row {i+1}:")
            for j in range(n):
                A[i, j] = np.double(input(f"A[{i+1}, {j+1}]: "))
        print("Please enter the values of the column vector b: ")
        for k in range(n):
            b[k] = np.double(input(f"b[{k+1}]: "))
        print("Please enter the initial approximations for x0: ")
        for l in range(n):
            x_0[l] = np.double(input(f"x_0[{l+1}]: "))
        flag = True

    if flag:
        flag = False
        while not flag:
            tol = float(input("Please input a value for the tolerance (tol): "))
            if tol <= 0.0:
                print("TOL must be a positive value.")
            else:
                flag = True
        flag = False
        while not flag:
            N = int(input("Please enter a value for the maximum number of iterations (N): "))
            if N <= 0:
                print("N0 must be a positive value.")
            else:
                flag = True
    return flag, n, A, b, x_0, tol, N

def main():
    flag = False
    n = 0
    A = np.empty(0)
    b = np.empty(0)
    x_0 = np.empty(0)
    tol = 0.0
    N = 0

    flag, n, A, b, x_0, tol, N = inp(flag, n, A, b, x_0, tol, N)
    x = np.zeros(n)

    print("This is the Gauss-Seidel Iterative Method.")

    if flag:
        k = 1
        while k <= N and flag:
            for i in range(n):
                sum1 = 0.0
                sum2 = 0.0
                for j in range(i):
                    sum1 += A[i, j] * x[j]
                for l in range(i+1, n):
                    sum2 += A[i, l] * x_0[l]
                x[i] = (-sum1 - sum2 + b[i]) / A[i, i]
            print(f"Iteration {k}:")
            for i in range(n):
                print(f"x[{i+1}] = {x[i]}")

            norm = abs(x[0] - x_0[0])
            for i in range(n):
                if abs(x[i] - x_0[i]) > norm:
                    norm = abs(x[i] - x_0[i])
            if norm < tol:
                print(f"The procedure was successful after {k} iterations.")
                print(f"||x - x_0|| = {norm}, < TOL = {tol}")
                flag = False

            if flag:
                k += 1
                for i in range(n):
                    x_0[i] = x[i]

    if flag:
        print("The procedure was not successful.")
        flag = False

if __name__=="__main__":
    main()
```
